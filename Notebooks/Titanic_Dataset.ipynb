# Quick Look at the Data

```python
import sklearn.datasets as datasets
import pandas as pd
import numpy as np

df = pd.read_csv('../Datasets/Titanic/train.csv')
df.drop(['Name', 'Ticket', 'Cabin', 'PassengerId'], axis=1, inplace=True)
df_features = df.drop(['Survived'], axis=1)
df_target = df['Survived']

print(df_features.head())
print('\n')
print(df_features.info())
```

Observations:
    - 'Pclass', 'Sex', and 'Embarked' are categorical features while the rest are numeric 
    - 'Age' and 'Embarked' have null values

```python
print(df_features['Pclass'].value_counts())
print('\n')
print(df_features['Sex'].value_counts())
print('\n')
print(df_features['Embarked'].value_counts())
```

```python
df_features.describe()
```

```python
import matplotlib.pyplot as plt
df_features.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
#Split the dataset into training and test data
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)
train_set
```

```python
#Looking at correlation
corr_matrix = train_set.corr()
corr_matrix
```

```python
import seaborn as sns
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="Blues")
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Imputer

df_features_num = train_set.drop(['Pclass', 'Sex', 'Embarked', 'Survived'], axis=1)
df_features_cat = train_set[['Pclass', 'Sex', 'Embarked']]

num_pipeline = Pipeline([
        ('imputer', Imputer(strategy="median")),
        ('std_scaler', StandardScaler()),
    ])

df_train_num = num_pipeline.fit_transform(df_features_num)
df_train_num = pd.DataFrame(df_train_num, columns=df_features_num.columns)

df_train_cat = pd.get_dummies(df_features_cat, columns=['Sex', 'Embarked'])

df_train_num = df_train_num.reset_index(drop=True)
df_train_cat = df_train_cat.reset_index(drop=True)
df_train = df_train_num.join(df_train_cat)
```

```python
from sklearn import tree
clf_dt = tree.DecisionTreeClassifier()
clf_dt = clf.fit(df_train, train_set['Survived'])
clf_dt.score(df_train,train_set['Survived'])
```

```python
from sklearn.ensemble import RandomForestClassifier
clf_rf = RandomForestClassifier()
clf_rf = clf.fit(df_train, train_set['Survived'])
clf_rf.score(df_train,train_set['Survived'])
```

```python
from sklearn.cross_validation import cross_val_score

print(np.mean(cross_val_score(clf_dt, df_train, train_set['Survived'], cv=10)))

print(np.mean(cross_val_score(clf_rf, df_train, train_set['Survived'], cv=10)))
```

```python
from sklearn.model_selection import GridSearchCV
# Create the parameter grid based on the results of random search 
param_grid = {
    'bootstrap': [True],
 'max_depth': [100],
 'max_features': [3],
 'min_samples_leaf': [4],
 'min_samples_split': [6],
 'n_estimators': [1000]
}
# Instantiate the grid search model
grid_search = GridSearchCV(estimator = clf_rf, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 2)
```

```python
# Fit the grid search to the data
grid_search.fit(df_train, train_set['Survived'])
grid_search.best_params_
```
